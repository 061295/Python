## Observations

```python
1. Most of the jobs accepted are for flexible working hours or Mon-Fri. 
2. Candidates between the age group 19-26 are more likely to accept the offer.
3. More female candidates have turned down the offer when compared to male candidates.
4. The acceptance ratio of single candidates is higher than the married candidates comparatively.
5. Permanent job types are likely to attract more candidates. 
6. The majority of the candidates with less than 50% salary hike have not accepted the offer.
7. Companies offering More than 12 days leaves are expected to have higher joining rate. 
8. Candidates prefer a Near working location with an average living cost. 
9. Acceptance ration is high if the companies provide Bonus. 
```
